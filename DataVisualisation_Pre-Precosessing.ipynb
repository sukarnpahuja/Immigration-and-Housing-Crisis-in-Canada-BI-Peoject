# 2. Population Growth Table

```python
#Prepocessing of population growth table
#Importing table for different population sources 
src_pop = pd.read_excel(xls, 'source - pop')
```

```python
# Importing total population table 
total_pop = pd.read_csv('/Users/sudhanshukharbanda/Documents/visualisation project/1710000901_databaseLoadingData (1).csv')
```

```python
src_pop.head()
```

```python
# Getting columns which we will be using for visualisation
src_pop=src_pop[['REF_DATE', 'Components of population growth', 'VALUE']]
```

```python
#Checking how the data will look
src_pop.head()
```

```python
src_pop.tail()
```

```python
# Getting last year from REF_DATE 
src_pop['year'] = src_pop['REF_DATE'].str.split('/').str.get(0)
```

```python
# using groupby to get desired population component with the sum values 
src_pop = src_pop.groupby(by=['year','Components of population growth']).agg({"VALUE":sum}).reset_index()
src_pop.head()
```

```python
pivot_pop = src_pop.pivot_table(index=['year'],columns=['Components of population growth'],values = ['VALUE']).reset_index()
pivot_pop.head()
```

```python
#joing the two columns
pivot_pop.columns = list(map("".join, pivot_pop.columns[:1] )) + list(map("_".join, [t[::-1] for t in pivot_pop.columns[1:] ]))
print(pivot_pop.columns)
```

```python
#getting the desired columns from the table
pivot_pop = pivot_pop[['year', 'Births_VALUE', 'Deaths_VALUE', 'Immigrants_VALUE']]
pivot_pop.head()
```

```python
total_pop.head()
```

```python
#Getting the Desired Columns
total_pop=total_pop[['REF_DATE','GEO','VALUE']]
total_pop.head()
```

```python
total_pop['REF_DATE'] = total_pop['REF_DATE'].astype(str)
total_pop = total_pop[total_pop['REF_DATE'].str.endswith('01')]
total_pop['REF_DATE'] = pd.to_datetime(total_pop['REF_DATE'], format='%Y-%m')
total_pop['year'] = total_pop['REF_DATE'].dt.strftime('%Y')
total_pop['VALUE_copy']=total_pop['VALUE']
```

```python
total_pop.head()
```

```python
#Getting population growth by subtracting population from the previous year
total_pop['VALUE_copy'] = total_pop['VALUE_copy'].shift(periods=1)
total_pop['change_pop'] = total_pop['VALUE']-total_pop['VALUE_copy']
total_pop['change_pop'] = total_pop['change_pop'].shift(periods=-1)
total_pop_1=total_pop.loc[total_pop['GEO'] == 'Canada']
```

```python
total_pop_1.tail()
```

```python
total_pop_1 = total_pop_1.drop(index=[84])
```

```python
total_pop_1.tail()
```

```python
total_pop_1=total_pop_1[['year','change_pop']]
```

```python
total_pop_1=total_pop_1.head().reset_index(drop =True)
```

```python
total_pop_1.head()
```

```python
#Merging the table with total population growth
result_dataset = pd.merge(pivot_pop, total_pop_1, on='year')
```

```python
result_dataset.head()
```

```python
### Getting house prices row
src_houseprc.head()
```

```python
#prepocessing the data to get the house price index
src_houseprc_h_price=src_houseprc.loc[src_houseprc['New housing price indexes'] == 'Total (house and land)']
src_houseprc_h_price['date'] = pd.to_datetime(src_houseprc_h_price['date'], format='%Y-%m-%d')
src_houseprc_h_price['year'] = src_houseprc_h_price['date'].dt.strftime('%Y-%m')
src_houseprc_h_price['year'] = src_houseprc_h_price['year'].astype(str)
src_houseprc_h_price = src_houseprc_h_price[src_houseprc_h_price['year'].str.endswith('12')]
src_houseprc_h_price.head().reset_index(drop=True)
```

```python
src_houseprc_h_price = src_houseprc_h_price[src_houseprc_h_price['year'] != '2001-12']
src_houseprc_h_price['year'] = pd.to_datetime(src_houseprc_h_price['year'], format='%Y-%m')
src_houseprc_h_price['year'] = src_houseprc_h_price['year'].dt.strftime('%Y')
src_houseprc_h_price=src_houseprc_h_price[['year','VALUE']]
```

```python
# Merging the above result table with house pricing table according to year
result_dataset = pd.merge(result_dataset, src_houseprc_h_price, on='year')
```

```python
src_houseprc_h_price.head()
```

```python
result_dataset_1 = result_dataset.rename(columns={'VALUE': 'House prices', 'change_pop': 'Population Growth',
                                                  'Immigrants_VALUE': 'Immigrants','Deaths_VALUE': 'Deaths',
                                                 'Births_VALUE': 'Births',})
result_dataset_1.head()
```

# 3. Dwelling Table

```python
src_dwelling = pd.read_excel(xls, 'source - dwelling')
```

```python
src_dwelling.head()
```

```python
src_dwelling=src_dwelling[['REF_DATE', 'Type of structure', 'VALUE']]
```

```python
#Getting Type of Structure
src_dwelling['REF_DATE'] = pd.to_datetime(src_dwelling['REF_DATE'], format='%Y-%m')
src_dwelling['year'] = src_dwelling['REF_DATE'].dt.strftime('%Y')
src_dwelling = src_dwelling.groupby(by=['year','Type of structure']).agg({"VALUE":sum}).reset_index()
src_dwelling.head()
```

```python
# Creating table using pivot to get the desired components
pivot = src_dwelling.pivot_table(index=['year'],columns=['Type of structure'],values = ['VALUE']).reset_index()
pivot.head()
```

```python
pivot.columns = list(map("".join, pivot.columns[:1] )) + list(map("_".join, [t[::-1] for t in pivot.columns[1:] ]))
print(pivot.columns)
```

```python
# Selecting the required columns that will be used for visualisation 
pivot = pivot[['year', 'Multiple dwelling building total_VALUE',
       'Single dwelling building total_VALUE', 'Total commercial_VALUE',
       'Total industrial_VALUE', 'Total institutional and governmental_VALUE']]
pivot.head()
```
